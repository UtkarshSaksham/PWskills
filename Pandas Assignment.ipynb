1. Mean of each and every column present in the dataframe

```python
l = []
for i in df1.columns: # Iterating over all columns
    l.append(df1.loc[:,i].mean()) # Finding mean of all columns
df1.loc['Mean'] = l
```

Question 8

```python
import datetime
import pandas as pd

date_entry1 = input('Enter a date in YYYY-MM-DD format') # Taking first date
year, month, day = map(int, date_entry1.split('-'))
date1 = datetime.date(year, month, day)

date_entry2 = input('Enter a date in YYYY-MM-DD format') # Taking second date
year, month, day = map(int, date_entry2.split('-'))
date2 = datetime.date(year, month, day)
 
d = date1-date2
 
print("Difference between two dates", pd.Timedelta(d))
```
