Entry/Exit app

```python
from flask import *
app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to our Application' 

@app.route('/exit')
def exit():
    return 'Thanks for visiting, visit again' 

@app.route('/<action>', methods = ['GET'])
def msg(action):
    if action == "entry":
        return redirect(url_for('welcome'))
    
    elif action == "exit":
        return redirect(url_for('exit'))

    else:
        return "Invalid Input"

if __name__ == '__main__':
    app.run(debug=True)
```
