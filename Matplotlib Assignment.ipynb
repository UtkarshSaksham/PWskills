Question 2

Scatter plot is the graph in which different pointers is projected along x and y axis.

```python
# Given
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Scatter Graph')
plt.scatter(x,y)
plt.show()
```

Question 3

Subplot function is used in python for plotting multiple graphs in a single cell.

```python
import matplotlib.pyplot as plt
```

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
```

```python
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
```

```python
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
```

```python
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, ax = plt.subplots()
ax.plot(x1, y1)
ax.plot(x2, y2)
ax.plot(x3, y3)
ax.plot(x4, y4)
```

Question 4

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent.

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company, profit, color ='#1ad683', 
        width = 0.4)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Bar Plot")
plt.show()
```

```python
plt.barh(company, profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Horizontal Bar Plot")
plt.show()
```

A box plot uses boxes and lines to depict the distributions of one or more groups of numeric data. Box limits indicate the range of the central 50% of the data, with a central line marking the median value.

```python
box1 = np.random.normal(100, 10, 200)
```

```python
box2 = np.random.normal(90, 20, 200)
```

```python
fig = plt.figure(figsize =(10, 7))
```

```python
data = [box1, box2]
```

```python
plt.boxplot(data)
plt.show()
```
