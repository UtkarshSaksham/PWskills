url_for() function is used for url building in flask. It takes two parameters one is name of the function another is the key value pair. On the basis of this function it will redirect to particular page.

```python
# python code to demonstrate the working of the url_for() function
from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route("/success")
def Transaction_Successful():
    return "Your transaction is successful"

@app.route("/failed")
def Transaction_Failed():
    return "Your transaction is failed"


@app.route("/transaction/<int:amount>")
def transaction(amount):
    result = ''
    if amount>0:
        result = 'Transaction_Successful'
    else:
        result = 'Transaction_Failed'
    return redirect(url_for(result))


if __name__=="__main__":
    app.run(host="0.0.0.0")
```
